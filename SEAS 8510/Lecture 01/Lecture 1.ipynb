```python
# This is an example of a vector in python.

import numpy as np

asList = [1,2,3]
asArray = np.array([1,2,3]) # 1D array
rowVec = np.array([ [1,2,3] ]) # row
colVec = np.array([ [1],[2],[3] ]) # column

print(f'asList: {np.shape(asList)}')
print(asList)


print(f'asArray: {asArray.shape}')
print(asArray)

print(f'rowVec: {rowVec.shape}')
print(rowVec)

print(f'colVec: {colVec.shape}')
print(colVec)
```

```python
# Scalar * Vector
s = 2
a = [3,4,5] # as list
b = np.array(a) # as np array
print(a*s)
print(b*s)
```

```python
# Scalar + Vector
s = 2
v = np.array([3,6])
s+v
```

```python
#Transpose
rowVec = np.array([ [1,2,3] ]) # row
colVec = rowVec.transpose()


print(f'rowVec: {rowVec.shape}')
print(rowVec)

print(f'colVec: {colVec.shape}')
print(colVec)
```

```python
# Magnitude
v = np.array([1,2,3,7,8,9])
v_dim = len(v)  # math dimensionality
v_mag = np.linalg.norm(v) # math magnitude, length, or norm

print("len()")
print(v_dim)

print("np.linalg.norm()")
print(v_mag)
```

```python
# Dot Product
v = np.array([1,2,3,4])
w = np.array([5,6,7,8])
np.dot(v,w)
```

```python
#Dot Product and angle
v = np.array([1,0])
w = np.array([1,1])
dot = np.dot(v,w)
angle = np.arccos(dot/(np.linalg.norm(v)*np.linalg.norm(w)))
angle = np.degrees(angle)
print(f"The dot product is {dot}. The angle is {angle:.2f} degrees",sep="\n")
```

```python
# Distributive
a = np.array([ 0,1,2 ])
b = np.array([ 3,5,8 ])
c = np.array([ 13,21,34 ])

# the dot product is distributive
res1 = np.dot( a, b+c )
print(res1)
res2 = np.dot( a,b ) + np.dot( a,c )
print(res2)
```

```python
# Correlation
def pearson_correlation(x, y):
 """
 Compute Pearson correlation coefficient between two vectors.
 """
 # Ensure numpy arrays
 x = np.array(x)
 y = np.array(y)
 # Mean-centering the vectors
 x = x - np.mean(x)
 y = y - np.mean(y)
 # Compute Pearson correlation coefficient
 correlation = np.dot(x, y) / (np.sqrt(np.dot(x, x)) * np.sqrt(np.dot(y, y)))
 return correlation
```
