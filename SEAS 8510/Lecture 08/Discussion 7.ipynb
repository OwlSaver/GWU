```python
import random
import pandas as pd

def generate_data(num_people):
  """
  Generates a DataFrame with age, wealth, and flexibility score.

  Args:
      num_people: The number of people to generate data for.

  Returns:
      A pandas DataFrame with age, wealth, and flexibility score columns.
  """
  data = []
  for _ in range(num_people):
    # Generate random age with normal distribution
    age = random.gauss(45, 22)
    age = max(0, min(age, 105))  # Ensure age is within range

    # Calculate linear increase in mean wealth with age
    mean_wealth = age * (5000000 / 105)

    # Generate wealth with normal distribution around the calculated mean
    wealth = random.gauss(mean_wealth, 1000000)  # Adjust standard deviation as desired
    #wealth = max(0, min(wealth, 5000000))  # Ensure wealth is within range
    #wealth = max(0, min(wealth, 5000000))  # Ensure wealth is within range

    # Calculate linear decrease in mean flexibility with age
    mean_flexibility = 5 - (age / 105) * 4  # Adjust slope as desired

    # Generate flexibility score with normal distribution around the calculated mean
    flexibility_score = random.gauss(mean_flexibility, 0.5)  # Adjust standard deviation as desired
    flexibility_score = max(1, min(flexibility_score, 5))  # Ensure flexibility score is within range

    person = {
      "Age": age,
      "Wealth": round(wealth),
      "Flexibility": round(flexibility_score, 2)
    }
    data.append(person)

  df = pd.DataFrame(data)
  return df

# Example usage
num_people = 1000
df = generate_data(num_people)

# Print the DataFrame (optional)
print("First five rows")
print(df.head())  # View the first few rows
print("")
print("Describe")
print(df.describe())
print("")
print(f"Covariance of age and wealth: {df['Age'].cov(df['Wealth'])}.")
print(f"Covariance of age and flexibility: {df['Age'].cov(df['Flexibility'])}.")
print(f"Covariance of wealth and flexibility: {df['Wealth'].cov(df['Flexibility'])}.")
```
