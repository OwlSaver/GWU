```python
import numpy as np
import matplotlib.pyplot as plt

def cdf(x):
  """
  This function defines a CDF for errors in a manufacturing process.
  - Low errors (below 1): Increase gradually.
  - Medium errors (between 1 and 3): Increase rapidly.
  - High errors (above 3): Level off towards 1.
  """
  if x < 0:
    return 0
  elif x < 1:
    return x**2 / 2
  elif x < 3:
    return 0.5 + (x - 1) * 0.75  # Steeper increase for medium errors
  else:
    return 1 - (x - 3)**2 / 4  # Levels off for high errors

# Sample size (large for smoother PDF)
samples = 10000

# Generate random values between 0 and 1
random_vals = np.random.rand(samples)

# Apply the CDF to get error values
error_vals = cdf(random_vals)

# Define x values for plotting the CDF smoothly
x_plot = np.linspace(min(error_vals), max(error_vals), 1000)
cdf_vals_plot = [cdf(val) for val in x_plot]

# Calculate PDF using kernel density estimation
plt.hist(error_vals, bins=20, density=True, edgecolor='black', alpha=0.7)
plt.plot(x_plot, cdf_vals_plot, label='CDF')
plt.xlabel('Error Value (X)')
plt.ylabel('Probability Density/Cumulative Probability')
plt.title('PDF and CDF for Manufacturing Errors')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cdf(x):
  """
  This function defines the cumulative distribution function (CDF).
  - Exponential distribution with parameter lambda = 2
  """
  if x < 0:
    return 0
  else:
    return 1 - np.exp(-2 * x)  # Exponential CDF with lambda = 2
```

```python
def generate_random_variable(cdf):
  """
  This function generates random variables from the provided CDF using inverse transform sampling.
  """
  # Sample size
  samples = 1000

  # Generate random values between 0 and 1
  random_vals = np.random.rand(samples)

  # Find the inverse CDF (may require numerical methods for complex CDFs)
  inverse_cdf = np.linalg.solve(lambda x: cdf(x) - random_vals, np.arange(0, 1, 0.001))

  return inverse_cdf
```

```python
x_plot = np.linspace(0, 5, 1000)  # Define x values for smooth plotting
cdf_vals_plot = [cdf(val) for val in x_plot]

# Calculate PDF using numerical differentiation
pdf_vals_plot = np.diff(cdf_vals_plot) / np.diff(x_plot)

plt.plot(x_plot, cdf_vals_plot, label='CDF')
plt.plot(x_plot[1:], pdf_vals_plot, label='PDF')  # Adjust x for PDF to avoid division by zero
plt.xlabel('Value (X)')
plt.ylabel('Probability')
plt.title('CDF and PDF of the Distribution')
plt.grid(True)
plt.legend()
plt.show()
```
