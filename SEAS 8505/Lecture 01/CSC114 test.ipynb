```python
# --------------------------------------------------------------------------------------
# 
# Show the representation and binary storage values for string, integer, and float
# variables. The built in bin() function works for integers. Functions need to be
# written for float and string variables.
# 
# --------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------
# 
# Converting floats to binary - based on IEEE format
# 
# --------------------------------------------------------------------------------------
import struct
def f2bin(float):
    return ''.join('{:0>8b}'.format(c) for c in struct.pack('!f', float))
# --------------------------------------------------------------------------------------
#
# Converting strings to binary. Assumes ASCII / UTF-8
#
# --------------------------------------------------------------------------------------
def c2bin(string):
    return ''.join(format(ord(i), '08b') for i in string)
# --------------------------------------------------------------------------------------
#
# Set the values for the variables
#
# --------------------------------------------------------------------------------------
A=1
a="abc"
B=123.47
NEW="OK"
# --------------------------------------------------------------------------------------
#
# Print the variable name, value, and value in binary
#
# --------------------------------------------------------------------------------------
print(f"The value of A is {A} and in binary {bin(A)}.")
print(f"The value of a is {a} and in binary {c2bin(a)}.")
print(f"The value of B is {B} and in binary {f2bin(B)}.")
print(f"The value of NEW is {NEW} and in binary {c2bin(NEW)}.")
```
